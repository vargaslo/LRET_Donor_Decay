# Define binary search algorithm

These functions serve separate purposes so they are defined a bit differently.

```python
%%px --local


# Binary search algorithm
# http://stackoverflow.com/questions/212358/binary-search-in-python

from bisect import bisect_left
from bisect import bisect_right


def binary_search(a, x, lo=0, hi=None):   # can't use a to specify default for hi
    hi = hi if hi is not None else len(a) # hi defaults to len(a)   
    pos = bisect_left(a,x,lo,hi)          # half-open intervals (c1, c2]
    return (pos if pos != hi else -1)     # don't walk off the end


def hist_bin_search(a, x, lo=0, hi=None):  # 'a' should be a list of bin_edges
    if x == a[-1]:
        return len(a)-2                       # include max value in last bin
    else:
        hi = hi if hi is not None else len(a) # hi defaults to len(a)   
        pos = bisect_right(a,x,lo,hi)         # half-open intervals [c1, c2)
        return (pos-1 if pos != hi else -1)   # don't walk off the end
```

# Define a random events function

This function picks a random time point from a exponential distribution.
The exponential distribution depends on the sum of all rates available to the system in its current state.

The function also picks a random transition from a uniform distribution.
The probabilty of the transition it picks depends on the ratio of its transition rate versus the sum of all rates available to the system in its current state.

```python
%%px --local


def get_transition_time(markov_state):
    
    #----------------------------------------
    # Get the list of rates available to the markov model
    # Loop over each donor to accumulate all outgoing rates
    rlist = []
    for i in range(4):
        position = i+1
    
        # If donor is in excited state, it may decay to ground state
        if markov_state[i]=='D*':
            rate = 'kd%s' % (position)
            rlist.append(rate)
            
            # If acceptor is not occupied, donor may transfer to acceptor
            if 'A*' not in markov_state:
                rate = 'kT%s' % (position)
                rlist.append(rate)
            
        # If acceptor is in excited state, it may decay to ground state
        if markov_state[i]=='A*':
            rate = 'ka'
            rlist.append(rate)

    #print rlist
    #----------------------------------------

    # list of rates converted to numeric
    rlistN = [eval(rate) for rate in rlist]

    Ri = np.cumsum(rlistN)
    RN = Ri[-1]
    
    # Pick a time
    u = 1 - np.random.uniform()
    dt = np.log(1/u)/RN
    
    # Pick a transition ... use half-bound intervals (a, b]
    u = 1 - np.random.uniform()
    which = binary_search(Ri, u * RN)
    
    out = rlist[which]
    # Select the transition from the list
    #out = info[which]

    return dt, out
```

# Generate random sequence of transitions

This function is where the system is actually simulated. This piece of the code could be modified to run on parallel systems as each instance is completely independent and embarrasingly parallel.



```python
%%px --local


def simulate_LRET(corenum):

    # Seed the random number generator
    np.random.seed()
    
    # Name the histogram
    histname = 'histogram_%s' % corenum
    hist_dict[histname] = np.zeros((len(histogram_bin_edges) - 1), dtype=int)

    # Start a timer
    tic=time.time()

    for i in range(int(num)):

        if 0:
        #if (i % (num/10.) == 0):
            toc = time.time()
            tictoc = int(toc - tic)
            tocstr = str(datetime.timedelta(seconds=tictoc))
            #print '%s s --- %d/100' % (tocstr, (1.*i/num*100.))

    
        state = ['D*','D*','D*','D*']
        t = 0
        for i in range(9):  # There are a maximum of 8 transitions for 4 donors

            #print "\nCurrent State - step %s" % i
            #print state
    
            if ('D*' in state) or ('A*' in state):
                dt, w = get_transition_time(state)
            
                t = t+dt

                # Loop over each donor
                # Modify the current state to reflect the transition
                rate = w
                if rate == 'ka':
                    state[state.index('A*')] = 'A'
                
                    # Output list of transition times
                    # out_A.append(t)
                    # Or build histogram on the fly?
                    bin_index = hist_bin_search(histogram_bin_edges, t)
                    hist_dict[histname][bin_index] += 1

                for i in range(4):
                    if (rate == 'kT%s' % (i+1)):
                        state[i] = 'A*'
                    if (rate == 'kd%s' % (i+1)):
                        state[i] = 'D'
            else:
                break


    return hist_dict


# This can be split into separate processes
#simulate_LRET(1)
```

```python
# Assemble all the histograms in hist_dict to obtain total
histogram_bin_edges = np.arange(0, .1, 1e-6)
histogram_bin_centers = 0.5 * (histogram_bin_edges + np.roll(histogram_bin_edges, -1))[0:-1]
histogram_total = np.zeros(len(histogram_bin_centers), dtype=int)
for i in result:
    for k,v in i.items():
        histogram_total = histogram_total+v
```

# Define model functions to use for parameter fitting

```python
if 'clients' in locals():
    numcores = len(clients.ids)
else:
    numcores = 1
    
# Fit to model function
def func1(x, a1, r1, r2, r3, r4):
    delta_t = 1e-6  # bin width of histogram data
    y = ((num * numcores) * delta_t * a1 *
         (r1 * np.exp(-(r1 + kd1) * x) + 
          r2 * np.exp(-(r2 + kd2) * x) +
          r3 * np.exp(-(r3 + kd3) * x) +
          r4 * np.exp(-(r4 + kd4) * x)))
    return y


# This function uses the known rates and coefficient
def func0(x, a1):
    a1 = 1
    y = func1(x, a1, kT1, kT2, kT3, kT4)
    return y


func_labels = {'func0': 'Heyduk Equation - Actual Values',
               'func1': 'Heyduk Fit - Fitted Values'}
```
